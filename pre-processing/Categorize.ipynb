### Transfering frames with annotations to another folder

```python
def transfer_labelled_file(annotation_folder, frame_folder):
    files_move_count = 0
    for frame_folders in os.listdir(frame_folder):
        print("Creating a new folder for labelled images for Folder " + frame_folders)
        new_folder = "labelledframes"
        labelled_frame_folder = os.path.join(frame_folder,frame_folders,new_folder)
        #making folder
        try:
            #Creating a folder
            if not os.path.exists(labelled_frame_folder):
                os.makedirs(labelled_frame_folder)
        #Raise an error if folder is not created
        except OSError:
            logging.error('Error creating directory')
    for annotation_folders in os.listdir(annotation_folder):
        print("Looking through annotation folder " + annotation_folders)
        files_infolder_move_count = 0
        for annotation_files in os.listdir(os.path.join(annotation_folder,annotation_folders)):
            annotation = os.path.join(annotation_folder,annotation_folders,annotation_files)
            if os.stat(annotation).st_size == 0:
                continue
            else:
                annotation_full_name = os.path.basename(annotation)
                name = os.path.splitext(annotation_full_name)[0]
                frame_file = os.path.join(frame_folder,annotation_folders,name + ".png")
                if os.path.exists(frame_file) == False :
                    continue
                else:
                    destination_folder = os.path.join(frame_folder,annotation_folders, new_folder)
                    shutil.move(frame_file,destination_folder )
                    files_move_count += 1
                    files_infolder_move_count += 1
        print("Total Number of Frames moved in folder " + annotation_folders + " is",files_infolder_move_count)
    print("Total Number of frames moved in Data Source is", files_move_count)
                
        
```

```python
#transfer labelled images in data source 1
Annotation_folder = "./Video_Dataset/Data_Source1/annotation/Annotation_folder/"
Frame_folder = "./Video_Dataset/Data_Source1/Frame/"
transfer_labelled_file(Annotation_folder, Frame_folder)
```

### Transfering frames with desired bbox annotation into another folder

```python
def transfer_labelled_bbox(annotation_folder, frame_folder):
    files_move_count = 0
    for frame_folders in os.listdir(frame_folder):
        print("Creating a new folder for bbox images for folder " + frame_folders)
        new_folder = "bbox in range"
        bbox_frame_folder = os.path.join(frame_folder, frame_folders, "labelledframes", new_folder)
        #making folder
        try:
            #Creating a folder
            if not os.path.exists(bbox_frame_folder):
                os.makedirs(bbox_frame_folder)
        #Raise an error if folder is not created
        except OSError:
            logging.error('Error creating directory')
    for annotation_folders in os.listdir(annotation_folder):
        print("Looking through annotation folder " + annotation_folders)
        files_infolder_move_count = 0
        for annotation_files in os.listdir(os.path.join(annotation_folder,annotation_folders)):
            annotation = os.path.join(annotation_folder, annotation_folders, annotation_files)
            if os.stat(annotation).st_size == 0:
                continue
            elif os.path.splitext(annotation)[1].lower() in [".txt"]:
                valid = True
                f = open(annotation, "r")
                area = []
                for lines in f:
                    words = lines.split()
                    bbox_area = int(words[6])
                    area.append(bbox_area)
                f.close()
                minimum, maximum = 2700, 4800
                for items in area:
                    if items < minimum or items > maximum :
                        valid = False
                        break
                    else:
                        continue
                if valid == True:
                    annotation_full_name = os.path.basename(annotation)
                    name = os.path.splitext(annotation_full_name)[0]
                    frame_file = os.path.join(frame_folder,annotation_folders, "labelledframes" ,name + ".png")
                    if os.path.exists(frame_file) == False :
                        continue
                    else:
                        destination_folder = os.path.join(frame_folder,annotation_folders, "labelledframes", new_folder)
                        shutil.move(frame_file, destination_folder)
                        files_move_count += 1
                        files_infolder_move_count += 1
        print("Total Number of Frames moved in folder " + annotation_folders + " is",files_infolder_move_count)
    print("Total Number of frames moved in Data Source is", files_move_count)
                        
                
        
```

### Checking if folder contain images of not desired resolution/size

```python
def check_non_HD_images(frame_folder):
    for frame_folders in os.listdir(frame_folder):
        #print("Looking through frame folder " + frame_folders)
        valid = True
        width, height = 1920, 1080
        labelled_frames_dir = os.path.join(frame_folder, frame_folders, "labelledframes")
        for images in os.listdir(labelled_frames_dir):
            img_dir = os.path.join(labelled_frames_dir, images)
            if os.path.splitext(img_dir)[1].lower() in ['.png', '.jpg', '.jpeg'] and valid == True:
                img = PIL.Image.open(img_dir)
                wid, hgt = img.size
                if wid < width or hgt < height:
                    valid = False
                    break
                else:
                    continue
            elif os.path.isdir(img_dir) == True and valid == True:
                for image in os.listdir(img_dir):
                    img_full_dir = os.path.join(img_dir,image)
                    imgs = PIL.Image.open(img_full_dir)
                    wid, hgt = imgs.size
                    if wid != width or hgt != height:
                        valid = False
                        break
                    else: 
                        continue
            else:
                continue
        if valid == False:
            print(frame_folders + " contain non HD Images")
```
