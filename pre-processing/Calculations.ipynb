### Counting the number of Frames in a Data Source

```python
def file_counter(frame_folder):
    file_count = 0
    for file in os.listdir(Frame_folder):
        file_count_for_folder = 0
        print("Calculating for total frames in " + file)
        for frame_images in os.listdir(os.path.join(Frame_folder, file)):
            if file == ".ipynb_checkpoints" :
                continue
            elif os.path.splitext(frame_images)[1].lower() in ['.png', '.jpg', '.jpeg']:
                file_count += 1
                file_count_for_folder +=1
        print("Total Frame images in folder " + file + " is" ,file_count_for_folder)

    print("Total number of frames for Data Source is " ,file_count)
```

### Counting the number of labelled images in a Data Source

```python
def annotations_counter(annotation_folder):
    annotation_count = 0
    #no_annotation_count = 0
    for folder in os.listdir(annotation_folder):
        annotation_count_for_folder = 0
        #no_annotation_count_folder = 0
        print("Calculating the total number of labeled images in folder " + folder)
        for annotations in os.listdir(os.path.join(annotation_folder,folder)):
            annotation_file = os.path.join(annotation_folder,folder,annotations)
            if os.stat(annotation_file).st_size == 0:
                continue
                #no_annotation_count_folder += 1
                #no_annotation_count += 1
            else:
                annotation_count_for_folder += 1
                annotation_count += 1
        
        print("Total labelled images in folder " + folder + " is", annotation_count_for_folder)
        #print("Total non-labelled images in folder " + folder + " is", no_annotation_count_folder)
    print("Total Labelled images in Data Source is", annotation_count)
    #print("Total non-Labelled images in Data Source is", no_annotation_count)
        
                
```

### Counting the number of labelled images that contain bbox within the desired area

```python
def bboxcounter(Annotation_folder):
    annotation_count = 0
    for folder in os.listdir(Annotation_folder):
        annotation_count_folder = 0
        print("Calculating the total number of labeled images with the desired bbox in folder " + folder)
        for annotations in os.listdir(os.path.join(Annotation_folder,folder)):
            annotation_file = os.path.join(Annotation_folder,folder,annotations)
            if os.stat(annotation_file).st_size == 0:
                continue
            else:
                valid = True
                f = open(annotation_file, "r")
                area = []
                for lines in f:
                    words = lines.split()
                    bbox_area = int(words[6])
                    area.append(bbox_area)
                f.close()
                minimum, maximum = 2700 , 4800
                for items in area:
                    if items < 2700 or  items > 4800:
                        valid = False
                        break
                    else:
                        continue
                if valid == True :
                    print(annotation_file)
                    annotation_count += 1
                    annotation_count_folder += 1
        print("Total labelled images with desired bbox in folder " + folder + " is", annotation_count_folder)
    print("Total Labelled images with desired bbox in Data Source is", annotation_count)
        
        
```

### Counting the number of frames that have been manually deleted

```python
def number_of_deleted(Frame_folder, Output_folder):
    count = 0
    for folder in os.listdir(Frame_folder):
        File_count = 0
        labelledframes = 'labelledframes'
        bbox = 'bbox in range'
        annotated_folder = os.path.join(Frame_folder, folder, labelledframes, bbox)
        if os.path.exists(annotated_folder) == False:
            continue
        else:
            for frames in os.listdir(annotated_folder):
                frame_file = os.path.join(annotated_folder, frames)
                name = os.path.splitext(frames)[0]
                output_file = os.path.join(Output_folder, folder, labelledframes, bbox, name + '.png')
                if os.path.exists(output_file) == False:
                    print(os.path.join(folder, labelledframes, bbox, frames) + ' has been deleted')
                    count += 1
                    File_count += 1
        print('Total number of images that have been deleted in folder ' + folder + ' are', File_count)
    print('Total number of images that have been deleted are ', count)
            
            
```

```python
Frame_folder = './Video_Dataset/Data_Source3/Frame/'
Output_folder = './Video_Dataset/Data_Source3/Outputs/'
number_of_deleted(Frame_folder, Output_folder)
```
